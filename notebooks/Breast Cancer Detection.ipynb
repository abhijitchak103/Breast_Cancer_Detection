Visualizing some mean features

```python
def plot_feature(col):
    """
    Plots the scatterplot and histplot of mean features
    """
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10,6))
    sns.scatterplot(x='id', y = col, data=data_df, ax=ax1, hue='diagnosis')
    sns.histplot(x= col, data=data_df, hue='diagnosis')
    plt.show()
```

```python
plot_feature('radius_mean')
```

```python
plot_feature('perimeter_mean')
```

```python
from sklearn.preprocessing import LabelEncoder
data_df['diagnosis'] = LabelEncoder().fit_transform(data_df.diagnosis)
data_df.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import roc_curve, roc_auc_score, f1_score, recall_score, precision_score
```

```python
X = data_df.drop(columns=['id', 'diagnosis'], axis=1)
y = data_df['diagnosis']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
```

```python
clf = LogisticRegression(max_iter=10000)
```

```python
clf.fit(X_train, y_train)
```

```python
y_pred = clf.predict(X_test)
```

```python
recall_score(y_test,y_pred)
```

```python
precision_score(y_test, y_pred)
```

```python
f1_score(y_test, y_pred)
```

```python
roc_auc_score(y_test, y_pred)
```
